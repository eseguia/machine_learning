```python
import pandas as pd
import math
import numpy as np
from sklearn import preprocessing, model_selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from matplotlib import style
import sys
style.use('ggplot')
```

```python
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

def handle_non_numerical_data(df):
    columns = df.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]

        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x+=1

            df[column] = list(map(convert_to_int, df[column]))

    return df
df.fillna(-99999, inplace = True)
df = handle_non_numerical_data(df)
df_test.fillna(-99999, inplace = True)
df_test = handle_non_numerical_data(df_test)

without_sales = pd.DataFrame.copy(df)
without_sales.drop(columns='SalePrice', inplace = True)
df
```

```python
X = without_sales.astype(float)
#X = preprocessing.scale(X)

y = np.array(df['SalePrice'])
print(X)
```

```python
#X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2)

X_train = X
y_train = y

example_measures = np.array(df_test)
example_measures = example_measures.reshape(len(example_measures), -1)

X_test = example_measures
#y_test = y[-test_size:]


clf = LinearRegression(n_jobs = -1)  # n_jobs = -1 runs as many jobs as your processor can handle

clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_train[-10:], y_train[-10:]) #test accuracy
print(accuracy)
```

```python
for X,y in zip(X_test,y_test):
    print(f'Model: {clf.predict([X])}, Actual: {y}')
```
